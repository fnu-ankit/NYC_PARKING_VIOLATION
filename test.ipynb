```python
import pandas as pd
import duckdb
```

```python
db_nyc_parking_violations = 'data/nyc_parking_violations.db'

sql_query_show_tables = """
    SHOW tables
    """

sql_query_import_1 = """
CREATE OR REPLACE TABLE parking_violation_codes AS
Select *
From read_csv_auto(
    'data/ParkingViolationCodes.csv'
    ,normalize_names=True
)
"""

sql_query_import_2 = """
CREATE OR REPLACE TABLE parking_violations_2024 AS
Select *
From read_csv_auto(
    'data/Parking_Violations_Issued_-_Fiscal_Year_2024.csv'
    ,normalize_names=True
)
"""
```

```python
def query_db(db, sql_query):

    with duckdb.connect(db) as con:
        display(con.sql(sql_query).df())

query_db(db_nyc_parking_violations, sql_query_show_tables)
```

```python
#dev
db_nyc_parking_violations = 'data/nyc_parking_violations.db'

with duckdb.connect(db_nyc_parking_violations) as con:
    con.sql(sql_query_import_1)
    con.sql(sql_query_import_2)
```
